```python
# Importar el módulo pyplot con el alias plt
import matplotlib.pyplot as plt
# Crear la figura y los ejes
fig, ax = plt.subplots()
# Dibujar puntos
ax.scatter(x = [3, 4, 5,6], y = [4, 3, 2, 1])
# Guardar el gráfico en formato png
plt.savefig('diagrama-dispersion.png')
# Mostrar el gráfico
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = [[3], [4], [5], [6]]
y = [[4.], [3.], [2.], [1.]]

plt.scatter(x, y, label = 'scatter') # Dibuja un diagrama de dispersión

linreg = LinearRegression()
linreg.fit(x, y)  # Ajuste de regresión
y_pred = linreg.predict(x) # Ingrese los datos 

plt.plot(x, y_pred, 'r', label = 'plot')  # Dibuja un gráfico de líneas
plt.legend()
plt.show()
```
