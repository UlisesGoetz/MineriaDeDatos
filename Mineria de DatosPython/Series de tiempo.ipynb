```python
import matplotlib.pyplot as plt #implementamos la libreria
```

```python
fig, ax = plt.subplots() #Creamos el grafico
years = ['1991', '1992', '1993', '1994', '1995', '1996'] #Damos valores al vector de las abscisas 
Ordenadas = {'Contaminacion':[4620,4910,5490,5730,5990,6216]} #Vector o vectores de las ordenadas
ax.plot(years, Ordenadas['Contaminacion'], marker = 'o')#Ordenamos los vectores en la grafica
plt.show()#Mostramos el resultado
```
